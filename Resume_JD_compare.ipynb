Tokenize the resume and job description texts to extract individual words.

Create sets of unique words (keywords) from both the resume and job description.

Compare the sets to find common keywords.

Calculate the percentage of keywords in your resume that match the ones in the job description.


```python
# Define a function to calculate percentage match for each label
def calculate_label_match(resume_doc, jd_doc):
    
    jd_labels = set(ent.label_ for ent in jd_doc.ents)
    label_matches = {}

    # Iterate over each label
    for label in jd_labels:
        
        jd_tokens = set(token.text.lower() for ent in jd_doc.ents if ent.label_ == label for token in ent)
        
        resume_tokens = set(token.text.lower() for token in resume_doc if not token.is_stop and not token.is_punct)

        common_keywords = resume_tokens.intersection(jd_tokens)

        # Calculate percentage match
        if len(jd_tokens) > 0:
            percentage_match = (len(common_keywords) / len(jd_tokens)) * 100
        else:
            percentage_match = 0

        label_matches[label] = percentage_match

    return label_matches

```

```python
label_matches = calculate_label_match(doc2, doc1)

for label, percentage in label_matches.items():
    print(f"Percentage match for label '{label}': {percentage:.2f}%")
```
